#subsetting the columns

```python
df = df.drop(columns=['idx'])
df
```

#changing opening_date_clean column which was object to datetime

```python
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
```

#Displaying the top 10 category of product purchased in bar graph

```python
ax = df['product_category'].value_counts() \
    .head(10) \
    .plot(kind='bar', title='Top Categories of Product')
ax.set_xlabel('Product Category')
ax.set_ylabel('Count')
```

#Displaying histogram of Product Amount

```python
ax = df['product_amount'].plot(kind='hist', bins=20, title='Product Amount')
ax.set_xlabel('Product Amount(Rs.)')
plt.show()
```

#displaying KDE(Kernel Density Estimation) of Product Amount

```python
ax = df['product_amount'].plot(kind='kde', title='Product Amount(Rs.)')
ax.set_xlabel('Amount(Rs.)')
```

#Top 10 Mechants with highest amount of transaction

```python
df.groupby('merchant_name')['product_amount'].sum().sort_values(ascending=False).head(10) \
        .plot(kind='bar', title='Top Merchants')
plt.show()
```

#Top 10 Active User

```python
df.groupby('user_id')['product_amount'].sum().sort_values(ascending=False).head(10)
```

#Displaying correlation between different features

```python
df_corr = df[['product_amount', 'transaction_fee', 'cashback', 'loyalty_points']].dropna().corr()
df_corr
```

#Displaying the correlation in heatmap

```python
sns.heatmap(df_corr, annot=True)
```

#Avreage amount of transaction by location

```python
ax = df.groupby('location')['product_amount'] \
    .agg(['mean','count']) \
    .sort_values('mean') ['mean']\
    .plot(kind='barh', figsize=(8,5), title='Average Transaction by Location')
ax.set_xlabel('Transaction Amount')
plt.show()
```

#Number of transactions in month

```python
df['transaction_date'].dt.to_period('M').value_counts().sort_index()
```

#Scatterplot of product_amount vs Cashback with respect to Payment Method

```python
sns.scatterplot(data=df, x='product_amount', y='cashback', hue='payment_method')
plt.title('Product Amount vs. Cashback by Payment Method')
plt.show()
```

#Integrating plotly

```python
import plotly.express as px
import plotly.graph_objects as go
```

#Displaying scatterplot of Product Amount vs Cashback

```python
fig1 = px.scatter(df, x='product_amount', y='cashback', color='payment_method',title='Product Amount vs Cashback by Payment Method',
                  hover_data=['merchant_name', 'transaction_date'])
fig1
```

#Top 10 merchants by total product amount

```python
top_merchants = df.groupby('merchant_name')['product_amount'].sum().nlargest(10).reset_index()
fig2 = px.bar(top_merchants, x='merchant_name', y='product_amount',
              title='Top 10 merchants by transaction amount', labels={'product_amount':'Total product amount'})
fig2
```

#Converting daily transaction to monthly and displaying number of Monthly transaction in Line graph

```python
df['Month'] = df['transaction_date'].dt.to_period('M').astype(str)
monthly_counts = df['Month'].value_counts().sort_index()
fig3 = px.line(monthly_counts, x=monthly_counts.index, y=monthly_counts.values,
               labels={'x':'Month', 'y':'Transaction count'},
               title='Monthly Transaction Volume')
fig3
```

#creating a new feature Net Spent and its distribution by product category

```python
df['net_spent'] = df['product_amount']+df['transaction_fee']-df['cashback']
fig4 = px.box(df, x='product_category', y='net_spent', color='product_category',
              title='Net Spent by product category')
fig4
```

#Displaying line graph of total monthly transaction product amount

```python
monthly_transaction =  df.groupby('Month')['product_amount'].sum().reset_index()
fig5 = px.line(monthly_transaction, x='Month', y='product_amount',title='Monthly Total Product Amount')
fig5
```
